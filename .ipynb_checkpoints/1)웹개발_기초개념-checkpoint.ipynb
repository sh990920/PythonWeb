# 웹 개발의 의의
- 과거에는 백엔드(back-end)개발에 템플릿만 붙여서 웹 개발을 진행
    - 최근에는 각 기기의 화면 크기에 맞춰 알아서 조정되는 반응형 웹, 다양한 상호작용, 애니메이션 효과 등 사용자들의 이용경험을 향상시키기 위해 다양한 시도가 이루어지면서 프론트엔드와 백엔드의 구분이 명확해짐

## 웹의 구조
<img src="./img/web-dev-flow.png">

- 프론트엔드 : 각종 기능들을 구현
    - 데이터 처리에 대한 기능을 제외한 대부분의 기능
        - 좋아요
        - 이미지 위에 커서를 가져다 놓았을 때 이미지에 대한 정보 활성화
        - 이미지나 텍스트 등을 웹에서 편집하는 등 사용자와 상호작용을 하는 다양한 기능들
    - 퍼블리셔가 없다면 웹 페이지 구조 작성, 디자인까지도 진행
    
- 백엔드 : 데이터 처리에 관한 모든 기능을 수행
    - 데이터베이스와 연동되어 저장된 데이터를 찾거나 수정, 삭제하는 기능
    - 로그인/회원가입에 대한 요청을 받아 처리
    - 게시판에 글을 작성하고, 내 피드에 글들이 나타나게 하는 기능

- 프론트엔드와 백엔드는 일정한 규칙을 가지고 데이터를 주고 받음
    - 프론트엔드 : 사용자와 직접 맞닿아 있으며, 사용자에게 화면을 보여주고 무언가 행동을 하도록 유도
        - 이 작업을 위해서는 백엔드로부터 적절한 데이터를 받아와야함
        - 따라서 프론트엔드가 백엔드에게 직접 데이터를 요청
    
    - 백엔드 : 프론트에게 해당 요청에 맞는 응답
        - 응답의 결과 : 프론트엔드가 요청한 데이터 또는 에러 메세지
        
- 사용자는 웹 페이지에서 버튼을 누르거나 무언가를 입력하면 그 결과가 바로바로 받고 실제로 그 행동에 대한 결과가 갱신된 것으로 받아들임
    - 프론트엔드와 백엔드가 긴밀하게 소통해야 함

## 프론트엔드와 백엔드

### 프론트엔드
- 사용자가 접하고 경험하는 웹의 보이는 부분
- 기본적으로 웹 페이지가 구성되는 것과 관련된 다양한 작업들을 다룸

- 웹 프론트엔드의 기본 구성요소
    - HTML : 웹 페이지의 기본 뼈대
    - CSS : 꾸미기에 사용
    - JavaScript : 웹 페이지에 여러가지 동작을 할 수 있도록 도와줌
    - 그 외 다양한 응용 버전 도구들
        - HTML을 더 편하게 만들기 위한 템플릿 도구
        - CSS 프레임워크(Bootstrap)
        - 프론트엔드 프레임워크
            - 프로젝트 구조를 쉽게 관리하고 여러 기본 기능이 구현된 JavaScript 기반 프레임워크
                - React.js
                - Vue.js
                
- 과거에는 정적인 웹 페이지 개발이 주로 이루어졌지만 최근 프론트엔드 개발은 사용자의 서비스 이용 경험을 향상시키는 것까지 개발의 범위에 포함하고 있음
    - 페이지가 로드되는 시간을 최소화
    - 다양한 애니메이션
    - 웹 페이지 디자인 및 구성

- 사용자와 가장 가까운 개발 분야이기 때문에 최신 트렌드에 발맞춰 빠르게 공부하고 새로운 도구에 빠르게 익숙해질 필요가 있음

### 백엔드
- 데이터에 집중하여 보이지 않는 영역에서 이루어지는 일들을 구현
    - 예) 웹 페이지에서 게시글을 불러오거나 로그인을 하게 되는 경우, 사용자들은 그 결과만 보게 되고 내부적으로 어떤 데이터를 참조해서 가져오는지 알 수 없음

- 백엔드는 데이터베이스와 주로 연동하게 되는데 데이터베이스를 설계하고 설계된 데이터베이스를 바탕으로 데이터를 생성, 수정, 삭제, 조회 등의 기능을 구현하고 다루는 것이 백엔드의 매우 주요한 역할 중 하나

- 백엔드 개발자들의 주요 업무
    - 효율적인 데이터베이스 설계
    - 빠르게 데이터를 가져올 수 있는 기능 개발
    - 서비스에 장애가 발생하지 않도록 안정적인 백엔드 기능을 개발

- 프론트엔드에 비해 비교적 기술의 변화 속도가 느리기 때문에 기본기만 잘 갖춰 놓으면 새로운 도구를 쓰게 되더라도 금방 적용 가능
    - 기본기를 충실히 익히는 것이 매우 중요

### 풀스택
- 프론트엔드와 백엔드를 둘 다 포함하는 전체를 일컫는 단어
    - 두 분야를 모두 통달하는 것은 쉽지 않지만 프론트엔드나 백엔드 모두 각각 쉽게 개발할 수 있도록 만들어진 프레임워크나 서비스들이 많기 때문에 이를 적절히 잘 활용하여 프론트엔드부터 백엔드까지 개발하기도 함

# Django
- 파이썬 기반의 웹 풀스택 프레임워크
- Django만으로 웹 개발을 처음부터 끝까지 완성해낼 수 있음
- Django는 다른 프레임워크들에 비해 자유도가 낮음
    - 건드릴 수 있는 부분이 적음 -> 몇 가지 건드리지 않아도 웹 어플리케이션이 완성된다
    - 초보자들에게 적합한 프레임워크

- Django를 사용중인 곳
    - 인스타그램
    - 번개장터
    - 숨은고수
    - 헤이딜러
    - 라프텔

# 개발 패턴
- 만들어야 하는 개발 요소들에 대한 규격화된 양식
- 개발 프로젝트를 요약하자면 개발에 필요한 내용들을 몇 가지 단위로 나눠놓고, 하나하나 만든 후에 이것들을 연결하는 것
    - Django에서는 MTV패턴을 사용
